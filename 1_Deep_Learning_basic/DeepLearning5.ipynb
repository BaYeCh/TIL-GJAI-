```python
from keras import models, layers

# 출력층에 활성 함수를 따로 지정하지 않는데 회귀 모델의 경우 분류 모델과 달리
# 훈련 데이터와 유사한 값을 출력으로 내놓기 때문에 활성함수를 사용하지 않음
model = models.Sequential()
model.add(layers.Dense(30, activation='relu',input_dim=13))
model.add(layers.Dense(16, activation='relu'))
model.add(layers.Dense(8, activation='relu'))
model.add(layers.Dense(1))
```

```python
model.summary()
```

```python
X = df.iloc[:,:13]
y = df.iloc[:,13]
```

```python
from sklearn.model_selection import train_test_split
train_X, test_X, train_y, test_y = train_test_split(X, y)
```

```python
model.compile(loss='mse',optimizer='adam')
history = model.fit(train_X, train_y, epochs=200, batch_size=10)
```

```python
plt.plot(test_X, test_y, 'bo')
plt.plot(test_X, model.predict(test_X),'p',color='red')
plt.show()
```

```python
model.predict(test_X)
```

```python
import numpy as np
```

```python
from sklearn.metrics import mean_squared_error
pred = model.predict(test_X)
rmse = np.sqrt(mean_squared_error(test_y, pred))
print(rmse)
```

```python
df = pd.read_csv('pima-indians-diabetes.csv', header= 0)
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
plt.figure(figsize=(12,9))
plt.subplot(121)
plt.boxplot(df, showmeans=True, meanline=True,labels=df.columns)
plt.grid()
plt.subplot(122)
plt.boxplot(df['insulin'], showmeans=True, meanline=True, labels = ['insulin'])
plt.grid(axis='y')
plt.show()
```

```python
X = df.iloc[:,0:8]
y = df.iloc[:,8]
```

```python
train_X, test_X, train_y, test_y = train_test_split(X,y)
print(len(train_X),len(test_X), len(train_y), len(test_y) )
```

```python
model = models.Sequential()
model.add(layers.Dense(12,activation = 'relu', input_dim = 8))
model.add(layers.Dense(8,activation= 'relu'))
model.add(layers.Dense(1, activation= 'sigmoid'))
model.summary()
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics= ['acc'])
```

```python
model.fit(train_X, train_y, epochs = 200, batch_size = 10)
```

```python
model.predict(test_X)
```

```python
plt.plot(test_X, test_y, 'bo')
plt.plot(test_X, model.predict(test_X),'ro')
plt.show()
```

```python
prediction = model.predict(test_X)

prediction2 = []
for i in range(len(prediction)):
    if prediction[i] > 0.5:
        prediction2.append(1)
    else:
        prediction2.append(0)
```

```python
plt.plot(test_X, test_y, 'bo')
plt.plot(test_X, prediction2, 'ro')
plt.show()
```
