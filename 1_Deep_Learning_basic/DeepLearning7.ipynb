```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('wine.csv', header = None)
```

```python
X = df.iloc[:,:12]
y = df.iloc[:,12]
```

```python
train_X, test_X, train_y , test_y = train_test_split(X,y)
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model =Sequential()
model.add(Dense(30, activation = 'relu', input_dim =12))
model.add(Dense(16,activation = 'relu'))
model.add(Dense(8,activation='relu'))
model.add(Dense(1, activation = 'sigmoid'))
```

```python
model.compile(loss = 'binary_crossentropy', optimizer = 'adam',metrics = ['acc'])
history = model.fit(train_X, train_y, epochs = 200, batch_size = 200, validation_split=0.2)
```

```python
# Acc & Loss 확인
acc = history.history['acc']
val_acc = history.history['val_acc']
loss = history.history['loss']
val_loss = history.history['val_loss']
print(val_acc)
```

```python
epochs = range(1, len(acc)+1)
plt.figure(figsize= (12,9))
plt.plot(epochs, acc, 'blue',linestyle='--' ,label = 'Training Acc')
plt.plot(epochs, val_acc, 'red', label = 'Validation Acc')
plt.title('Training and Validation Accuracy')
plt.legend()
plt.show()
```

```python
model.evaluate(test_X,test_y)[1]
```
