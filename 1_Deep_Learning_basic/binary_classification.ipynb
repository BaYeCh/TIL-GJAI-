```python
def read_data(filename, encoding = 'cp949',start = 1):
    with open(filename, 'r', encoding = encoding) as f:
        data = [line.split('\t') for line in f.read().splitlines()]
        data = data[start:]
    return data
def write_data(data, filename, encoding = 'cp949'):
    with open(filename, 'w', encoding = encoding) as f:
        f.write(data)
```

# Tokenizing

```python
# 토큰화
from keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
import math
```

```python
max_words = 10000 # 데이터셋에서 가장 빈도 높은 10,000 개의 단어만 사용
                  # n-1개의 단어를 가져오는 데 이는 padding을 위한 단어 하나를 남겨 둠
maxlen = 20 # 20개 이후의 단어는 버려서 각 문장의 길이를 고정
            # 20보다 길면 뒤를 자르고, 작으면 0으로 채움
```

```python
tokenizer = Tokenizer(num_words=max_words) # 상위빈도 10,000개 단어를 추려내는 Tokenizer 객체 생성
tokenizer.fit_on_texts(train_text) # 전체 단어를 대상으로 인덱스 구축(빈도 기준)
word_index = tokenizer.word_index # 단어 인덱스 가져오기
```

```python
print('전체 {}개\n타입 : {}\nword_index : {}'.format(len(word_index), type(word_index), word_index))
```

### Data Sequencing

```python
# 텍스틀 숫자로 변환
data = tokenizer.texts_to_sequences(train_text)
print('data 0 : {}\n text 0 : {}'.format(data[0],train_text[0]))
```

# Data Padding

```python
data = pad_sequences(data, maxlen= maxlen)
print('data : {}\ndata 0 : {}\ndata 0의 길이 : {}'.format(data,data[0],len(data[0])))
```

# One-Hot Encoding

```python
sample = [[5,6,7],[8,9,10]]
# 10은 11번째에 들어감 따라서 11개의 공간을 가진 np.array를 만듦
arr = np.zeros((len(sample), 10+1))
for i, seq in enumerate(sample):
    # 리스트가 2개 i는 총 2회
    # 각 i에서 리스트의 number가 가리키는 곳을 1을 넣음
    arr[i, seq] = 1
arr
```

```python
def to_one_hot(sequences, dimension):
    results = np.zeros((len(sequences), dimension))
    for i, sequence in enumerate(sequences):
        results[i, sequence] = 1
    return results
```

```python
data = to_one_hot(data, dimension=max_words)
labels = np.array(train_senti).astype('float32')
```
