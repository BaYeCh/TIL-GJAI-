```python
from keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
max_words = 10000
maxlen = 16
```

```python
# tokenizing
tokenizer = Tokenizer(num_words= max_words)
tokenizer.fit_on_texts(train_emo_X)
tokenizer.word_index
```

```python
# sequencing
train_emo_X_sequencing = tokenizer.texts_to_sequences(train_emo_X)
train_emo_X_sequencing
```

```python
# padding
train_emo_X_padding = pad_sequences(train_emo_X_sequencing, maxlen=maxlen)
train_emo_X_padding
```

```python
# OHE
def to_one_hot(sequences, dimension):
    results = np.zeros((len(sequences),dimension))
    for i, sequence in enumerate(sequences):
        results[i, sequence] = 1
    return results
```

```python
train_emo_X_1hot = to_one_hot(train_emo_X_padding, dimension= max_words)
train_emo_X_1hot
```

```python
train_train_X, train_valid_X, train_train_y, train_valid_y = train_test_split(train_emo_X_1hot, train_emo_y_1hot, train_size = 0.7)
```

```python
from keras.models import Sequential
from keras.layers import Dense

model = Sequential()
model.add(Dense(64, activation = 'relu', input_shape=(max_words,)))
model.add(Dense(32, activation = 'relu'))
model.add(Dense(16, activation = 'relu'))
model.add(Dense(6, activation = 'softmax'))
model.summary()
```

```python
len(train_train_X)
```

```python
from keras.callbacks import EarlyStopping, ModelCheckpoint
model.compile(loss= 'categorical_crossentropy', optimizer = 'rmsprop', metrics = ['acc'])
```
